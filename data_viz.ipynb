```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load CSV data
df = pd.read_csv("sales.csv")
print("ðŸ‘‰ Dataset Preview:")
display(df.head())

# Step 3: Basic Statistics
print("\nðŸ‘‰ Summary Statistics:")
display(df.describe())

# Step 4: Plotting

# 4a. Bar Chart - Sales per Month
plt.figure(figsize=(8,5))
plt.bar(df['Month'], df['Sales'], color='skyblue')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# 4b. Line Plot - Sales vs Profit
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Profit'], marker='s', label='Profit')
plt.title("Sales & Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.show()

# 4c. Histogram - Sales Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=5, color='lightgreen', edgecolor='black')
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# 4d. Scatter Plot - Sales vs Profit
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=df, s=100, color='red')
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
